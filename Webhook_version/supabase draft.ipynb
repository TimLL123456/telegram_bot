### Import Library

```python
import os
from supabase import create_client
import pandas as pd

from dotenv import load_dotenv
load_dotenv()
```

### Set Supabase Config

```python
url= os.environ.get("SUPABASE_URL")
key= os.environ.get("SUPABASE_KEY")
supabase = create_client(url, key)
```

### Read Table

```python
response = (
    supabase.table("transactions")
    .select("*")
    .execute()
)
pd.DataFrame(response.data)
```

```python
response = (
    supabase.table("transactions")
    .select("*")
    .eq("description", "Monthly salary")
    .eq("amount", 10000)
    .execute()
)
pd.DataFrame(response.data)
```

### Insert Data to Table

```python
transactions = [
    {
        "user_id": 1002,
        "date": "2025-06-04",
        "category_id": 1,
        "description": "Monthly salary",
        "currency": "USD",
        "amount": 10000.00
    },
    {
        "user_id": 1002,
        "date": "2025-06-05",
        "category_id": 2,
        "description": "Grocery shopping",
        "currency": "USD",
        "amount": 150.50
    },
    {
        "user_id": 1002,
        "date": "2025-06-06",
        "category_id": 3,
        "description": "Netflix Subscription",
        "currency": "USD",
        "amount": 15.99
    }
]
```

```python
response = (
    supabase.table("transactions")
    .insert(transactions)
    .execute()
)

pd.DataFrame(response.data)
```

### Update

```python
response = (
    supabase.table("transactions")
    .update({"description": "Unknown"})
    .eq("transaction_id", 12)
    .execute()
)

eval_update = (
    supabase.table("transactions")
    .select("*")
    .execute()
)
pd.DataFrame(eval_update.data)
```

### Delete

```python
response = (
    supabase.table("transactions")
    .delete()
    .eq("transaction_id", 11)
    .execute()
)
```
